```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

```python
# Apne dataset ka path
data_path = "../mnt/data/laptop_data.csv"  # adjust if needed
df = pd.read_csv(data_path)
df.shape
```

```python
df.head()
```

```python
df.info()
print("\nMissing values per column:\n", df.isna().sum())
```

```python
print("Target column exists:", 'Price_INR' in df.columns)
df['Price_INR'].describe()
```

```python
df.describe(include='all').T
```

```python
# Histogram for target
sns.histplot(df['Price_INR'], kde=True)
plt.title("Laptop Price Distribution (INR)")
plt.show()
```

```python
num_cols = df.select_dtypes(include=[np.number]).columns.tolist()
corr = df[num_cols].corr()
plt.figure(figsize=(10,6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Numeric Feature Correlation")
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.boxplot(x='Brand', y='Price_INR', data=df)
plt.xticks(rotation=45)
plt.title("Brand vs Laptop Price")
plt.show()

sns.scatterplot(x='RAM', y='Price_INR', data=df)
plt.title("RAM vs Price")
plt.show()
```

```python
df.drop_duplicates(inplace=True)
df.to_csv('../artifacts/raw/laptop_raw_clean.csv', index=False)
print("Cleaned data saved!")
```
